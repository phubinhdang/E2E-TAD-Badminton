```python
import json

with open('./data/tad_detection.json') as f:
    d = json.load(f)
```

```python
pred_segs = []
for i, (video_id, segs) in enumerate(d['results'].items()):
    for seg in segs:
        start = seg['segment'][0]
        end = seg['segment'][1]
        label = seg['label']
        score = seg['score']
        pred_segs.append({'video_id': video_id,
                          'start': start,
                          'end': end,
                          'label': label,
                          'score': score})
```

```python
# ginting_axelsen_clips = [item for item in d.items() if item[0].startswith('ginting_axelsen')]
from dataclasses import dataclass 
@dataclass
class Clip:
    name: str
    duration_in_seconds: float
    start_time: float
```

```python
import json
clips = []
cumulative_duration = 0
match_name = "ginting_axelsen"
with open('./data/badminton_img10fps_info.json') as f:
    json_data = json.load(f)
    for name, features in json_data.items():
        if name.startswith(match_name):
            duration_in_seconds = features["feature_second"]
            print(name, duration_in_seconds)
            clip = Clip(name=name, duration_in_seconds=duration_in_seconds, start_time=round(cumulative_duration, 2))
            clips.append(clip)
            cumulative_duration += duration_in_seconds
```

```python
abs_starts = []
abs_ends = []
for _, row in df.iterrows():
    clip_info = next(c for c in clips if c.name == row['video_id'])
    seg_start = row['start']
    seg_end = row['end']
    if seg_start < 0:
        print(f"seg_start = {seg_start} is changed to 0")
        seg_start = 0
    if seg_end > clip_info.duration_in_seconds:
        print(f"seg_end = {seg_start} exceed clip len and is set to clip len")
        seg_end = clip_info.duration_in_seconds
    abs_starts.append(clip_info.start_time + seg_start)
    abs_ends.append(clip_info.start_time + seg_end)
    
df['abs_start'] = abs_starts
df['abs_end'] = abs_ends
```

```python
df = df.sort_values(by=['video_id', 'start'], ascending=[True, True])
df = df.reset_index(drop=True)
```

```python
df = df[['abs_start', 'abs_end']]
```

```python
df = df.rename(columns={'abs_start': 'start', 'abs_end': 'end'})
```

```python
df
```

```python
def ss_to_mmss(seconds):
    minutes = int(seconds // 60)
    seconds = int(seconds % 60)
    return f"{minutes:02}:{seconds:02}"


# Create a new column with the mm:ss format
df = df.assign(start_mmss=df['start'].apply(ss_to_mmss))
df = df.assign(end_mmss=df['end'].apply(ss_to_mmss))
```

```python
df
```

```python
merged_intervals = []

# Start with the first interval
current_start = df.loc[0, 'start']
current_end = df.loc[0, 'end']

# Iterate through the DataFrame rows
for i in range(1, len(df)):
    row_start = df.loc[i, 'start']
    row_end = df.loc[i, 'end']

    # If the current interval overlaps with the previous one, merge them
    if row_start <= current_end:
        current_end = max(current_end, row_end)  # Update the end of the merged interval
    else:
        # No overlap, add the previous interval and start a new one
        merged_intervals.append([current_start, current_end])
        current_start = row_start
        current_end = row_end

# Append the last interval
merged_intervals.append([current_start, current_end])

# Create a new DataFrame from the merged intervals
merged_df = pd.DataFrame(merged_intervals, columns=['start', 'end'])
merged_df = merged_df.assign(start_mmss=merged_df['start'].apply(ss_to_mmss))
merged_df = merged_df.assign(end_mmss=merged_df['end'].apply(ss_to_mmss))
merged_df
```
